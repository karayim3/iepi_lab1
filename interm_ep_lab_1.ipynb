```python
import numpy as np
import matplotlib as plt
```

```python
# wave equation: A * np.sin(omega * t + phi) + C = 0
# find amplitude: A = -C / (np.sin(omega * t + phi))
omega = 6.39 #rad/sec
time = np.linspace(10, 60, 6) #sec
phi = -3.29 #rad
C = -8.83 * 10**-4

amplitude = []
for t in time:
  amplitude.append(-C / (np.sin(omega * t + phi)))

plt.pyplot.plot(time, amplitude, label='data')

ln_amp = []
for a in amplitude:
  ln_amp.append(np.log(a))
# there is no values for t=10,20,30 seconds because those values of the amplitude are negative
plt.pyplot.plot(time, ln_amp, label='natural logs')
plt.pyplot.title("Comparing the Amplitudes of Our Data and its Natural Logarithms")
plt.pyplot.xlabel("Time (seconds)")
plt.pyplot.ylabel("Amplitude (radians)")
plt.pyplot.legend()
```
